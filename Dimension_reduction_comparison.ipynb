```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as  sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print (type(cancer))
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
# classified data
print(cancer['target'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# normalize the variance of all features to unit variance 
scalar = StandardScaler()
scalar.fit(df)
scaled_data = scalar.transform(df)
print(scaled_data)

```

```python
plt.figure(figsize=(8,6))
plt.scatter(scaled_data[:,0], scaled_data[:,1], c= cancer['target'])
plt.xlabel('0')
plt.ylabel('1')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(scaled_data[:,1], scaled_data[:,2], c= cancer['target'])
plt.xlabel('1')
plt.ylabel('2')
```

```python
# PCA 
from sklearn.decomposition import PCA
pca = PCA( n_components=2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c= cancer['target'])
plt.xlabel('first principal component -pca1')
plt.ylabel('second principal component - pca2')
```

```python
#FACTOR ANALYSIS

from sklearn.decomposition import FactorAnalysis
FA = FactorAnalysis(n_components = 2 ).fit_transform(scaled_data)
```

```python
plt.figure(figsize=(12,8))
plt.title('Factor Analysis Components')
plt.scatter(FA[:,0], FA[:,1], c = cancer['target'])
#plt.scatter(FA[:,1], FA[:,2])
#plt.scatter(FA[:,2],FA[:,0])
```

```python
#t-SNE

from sklearn.manifold import TSNE 
tsne = TSNE(n_components=2, n_iter=300).fit_transform(scaled_data)
```

```python
plt.figure(figsize=(12,8))
plt.title('t-SNE components')
plt.scatter(tsne[:,0], tsne[:,1] ,c=cancer['target'])
#plt.scatter(tsne[:,1], tsne[:,2])
#plt.scatter(tsne[:,2], tsne[:,0])
```

```python
#UMAP
import umap
umap_data = umap.UMAP(n_neighbors=5, min_dist=0.3, n_components=2).fit_transform(scaled_data)
```

```python
plt.figure(figsize=(12,8))
plt.title('Decomposition using UMAP')
plt.scatter(umap_data[:,0], umap_data[:,1],c=cancer['target'])
#plt.scatter(umap_data[:,1], umap_data[:,2])
#plt.scatter(umap_data[:,2], umap_data[:,0])
```
